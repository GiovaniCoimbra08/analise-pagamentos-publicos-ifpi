```python
arquivos = {
    "jan2026": "dados_brutos/pagamentos_jan2026.csv",
    "fev2026": "dados_brutos/pagamentos_fev2026.csv",
    "abr2026": "dados_brutos/pagamentos_abr2026.csv",
    "mai2026": "dados_brutos/pagamentos_mai2026.csv"
}

for mes, caminho in arquivos.items():
    df_temp = pd.read_csv(caminho)
    print(mes, df_temp.shape)
```

```python
lista_dfs = []

for mes, caminho in arquivos.items():
    df_temp = pd.read_csv(caminho)
    df_temp["mes_arquivo"] = mes
    lista_dfs.append(df_temp)

df = pd.concat(lista_dfs, ignore_index=True)

df.shape
```

```python
colunas_data = [
    "data_emissao",
    "mes_referencia",
    "data_extracao",
    "data_atualizacao_registro"
]

for coluna in colunas_data:
    df[coluna] = pd.to_datetime(df[coluna], errors="coerce")
```

```python
df[colunas_data].dtypes
```

```python
def converter_valor(serie):
    if pd.api.types.is_numeric_dtype(serie):
        return serie

    serie_texto = serie.astype(str).str.strip()

    tem_virgula = serie_texto.str.contains(",", regex=False, na=False).any()

    if tem_virgula:
        serie_texto = (
            serie_texto
            .str.replace(".", "", regex=False)
            .str.replace(",", ".", regex=False)
        )

    serie_texto = serie_texto.replace({
        "": pd.NA,
        "nan": pd.NA,
        "None": pd.NA
    })

    return pd.to_numeric(serie_texto, errors="coerce")


for coluna in colunas_valor:
    df[coluna] = converter_valor(df[coluna])
```

```python
resumo_mes = df_positivos.groupby("mes_arquivo").agg(
    qtd_pagamentos=("codigo_pagamento", "count"),
    valor_total=("valor_pagamento_convertido_rs", "sum"),
    valor_medio=("valor_pagamento_convertido_rs", "mean")
).reset_index()

resumo_mes["valor_total"] = resumo_mes["valor_total"].round(2)
resumo_mes["valor_medio"] = resumo_mes["valor_medio"].round(2)

resumo_mes
```

```python
ordem_meses = {
    "jan2026": 1,
    "fev2026": 2,
    "abr2026": 4,
    "mai2026": 5
}

resumo_mes["ordem_mes"] = resumo_mes["mes_arquivo"].map(ordem_meses)

resumo_mes = resumo_mes.sort_values("ordem_mes").drop(columns="ordem_mes")

resumo_mes
```

```python
top_favorecidos = df_positivos.groupby("favorecido").agg(
    qtd_pagamentos=("codigo_pagamento", "count"),
    valor_total=("valor_pagamento_convertido_rs", "sum"),
    valor_medio=("valor_pagamento_convertido_rs", "mean")
).reset_index()

top_favorecidos["valor_total"] = top_favorecidos["valor_total"].round(2)
top_favorecidos["valor_medio"] = top_favorecidos["valor_medio"].round(2)

top_favorecidos = top_favorecidos.sort_values(
    by="valor_total",
    ascending=False
)

top_favorecidos.head(10)
```

```python
favorecidos_top5 = top_favorecidos.head(5)["favorecido"].tolist()

analise_top_favorecidos = df_positivos[
    df_positivos["favorecido"].isin(favorecidos_top5)
].groupby([
    "favorecido",
    "grupo_despesa",
    "elemento_despesa"
]).agg(
    qtd_pagamentos=("codigo_pagamento", "count"),
    valor_total=("valor_pagamento_convertido_rs", "sum"),
    valor_medio=("valor_pagamento_convertido_rs", "mean")
).reset_index()

analise_top_favorecidos["valor_total"] = analise_top_favorecidos["valor_total"].round(2)
analise_top_favorecidos["valor_medio"] = analise_top_favorecidos["valor_medio"].round(2)

analise_top_favorecidos.sort_values(
    by="valor_total",
    ascending=False
).head(20)
```

```python
resumo_grupo = df_positivos.groupby("grupo_despesa").agg(
    qtd_pagamentos=("codigo_pagamento", "count"),
    valor_total=("valor_pagamento_convertido_rs", "sum"),
    valor_medio=("valor_pagamento_convertido_rs", "mean")
).reset_index()

resumo_grupo["valor_total"] = resumo_grupo["valor_total"].round(2)
resumo_grupo["valor_medio"] = resumo_grupo["valor_medio"].round(2)

resumo_grupo = resumo_grupo.sort_values(
    by="valor_total",
    ascending=False
)

resumo_grupo
```

```python
total_geral = df_positivos["valor_pagamento_convertido_rs"].sum()

resumo_grupo["participacao_percentual"] = (
    resumo_grupo["valor_total"] / total_geral * 100
).round(2)

resumo_grupo
```

```python
resumo_elemento = df_positivos.groupby(["grupo_despesa", "elemento_despesa"]).agg(
    qtd_pagamentos=("codigo_pagamento", "count"),
    valor_total=("valor_pagamento_convertido_rs", "sum"),
    valor_medio=("valor_pagamento_convertido_rs", "mean")
).reset_index()

resumo_elemento["valor_total"] = resumo_elemento["valor_total"].round(2)
resumo_elemento["valor_medio"] = resumo_elemento["valor_medio"].round(2)

resumo_elemento = resumo_elemento.sort_values(
    by="valor_total",
    ascending=False
)

resumo_elemento.head(15)
```

```python
resumo_elemento_top15 = resumo_elemento.head(15).copy()

total_positivos = df_positivos["valor_pagamento_convertido_rs"].sum()

resumo_elemento_top15["participacao_percentual"] = (
    resumo_elemento_top15["valor_total"] / total_positivos * 100
).round(2)

resumo_elemento_top15
```

```python
resumo_mes_grupo = df_positivos.groupby(["mes_arquivo", "grupo_despesa"]).agg(
    qtd_pagamentos=("codigo_pagamento", "count"),
    valor_total=("valor_pagamento_convertido_rs", "sum"),
    valor_medio=("valor_pagamento_convertido_rs", "mean")
).reset_index()

resumo_mes_grupo["valor_total"] = resumo_mes_grupo["valor_total"].round(2)
resumo_mes_grupo["valor_medio"] = resumo_mes_grupo["valor_medio"].round(2)

ordem_meses = {
    "jan2026": 1,
    "fev2026": 2,
    "abr2026": 4,
    "mai2026": 5
}

resumo_mes_grupo["ordem_mes"] = resumo_mes_grupo["mes_arquivo"].map(ordem_meses)

resumo_mes_grupo = resumo_mes_grupo.sort_values(
    by=["ordem_mes", "valor_total"],
    ascending=[True, False]
).drop(columns="ordem_mes")

resumo_mes_grupo
```

```python
pivot_mes_grupo = resumo_mes_grupo.pivot_table(
    index="mes_arquivo",
    columns="grupo_despesa",
    values="valor_total",
    aggfunc="sum",
    fill_value=0
).reset_index()

ordem_meses = {
    "jan2026": 1,
    "fev2026": 2,
    "abr2026": 4,
    "mai2026": 5
}

pivot_mes_grupo["ordem_mes"] = pivot_mes_grupo["mes_arquivo"].map(ordem_meses)

pivot_mes_grupo = pivot_mes_grupo.sort_values("ordem_mes").drop(columns="ordem_mes")

pivot_mes_grupo
```

```python
maiores_pagamentos = df_positivos[[
    "mes_arquivo",
    "data_emissao",
    "favorecido",
    "grupo_despesa",
    "elemento_despesa",
    "observacao",
    "valor_pagamento_convertido_rs"
]].sort_values(
    by="valor_pagamento_convertido_rs",
    ascending=False
)

maiores_pagamentos.head(15)
```
